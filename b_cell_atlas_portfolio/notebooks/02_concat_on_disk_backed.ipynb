```python
import anndata.experimental

# Paths to input files and output file
file1 = "path/to/file1"
file2 = "path/to/file2"
output_file = "path/to/output/file"

# Use concat_on_disk to merge the files
anndata.experimental.concat_on_disk(
    in_files=[file1, file2],  # List of input files
    out_file=output_file,    # Path to the output file
    axis=0,                  # Concatenate cells (rows)
    join="inner",            # Use only common genes across datasets
    label="source",          # Add a column in .obs indicating dataset origin
    keys=["file1", "file2"],    # Labels for the datasets
    max_loaded_elems=100000000  # Adjust memory usage (default is fine for most cases)
)

print(f"Merged dataset saved to {output_file}")
```
