```python
# Import statements
import scanpy as sc
import anndata as ad
import pandas as pd
import os
import gc
import numpy as np
import matplotlib.pyplot as plt
```

```python
adata = sc.read_h5ad("/home/sbadol2/python_data/scpc3/SCPCL000050_unfiltered_rna.h5ad")
```

```python
sc.pp.filter_cells(adata, min_counts=1000) 
sc.pp.filter_cells(adata, min_genes=200) 
```

```python
adata = adata[adata.obs['subsets_mito_percent'] < 20]
```

```python
print(f"Min value: {adata.X.min()}")
print(f"Max value: {adata.X.max()}")
```

```python
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
print("Normalization done")
```

```python
sc.pp.highly_variable_genes(adata)
print("Find variable features done")
```

```python
sc.pp.scale(adata)
print("Scaling data done")
```

```python
sc.pp.pca(adata)
print("PCA done")
```

```python
sc.pp.neighbors(adata, n_pcs=50)
print("Find neighbors done")
```

```python
sc.tl.leiden(adata, resolution=1.0, key_added='seurat_clusters')
print("Find clusters done")
```

```python
sc.tl.umap(adata)
```

```python
sc.pl.umap(adata, color="seurat_clusters", frameon=False)
```

```python
sc.tl.tsne(adata, use_rep="X_pca", n_pcs=50)
sc.pl.tsne(adata, color="seurat_clusters", frameon=False)
```

```python
import scanpy.external as sce

# Run TriMap
sce.tl.trimap(adata)
sce.pl.trimap(adata, color="seurat_clusters", frameon=False)
```

```python
sc.tl.leiden(adata)
sc.tl.paga(adata)
sc.pl.paga(adata, color="seurat_clusters", frameon=False)
```

```python
sc.tl.draw_graph(adata, layout='fa')
print("Force-directed graph (FA) done")
sc.pl.draw_graph(adata, layout='fa', color = "seurat_clusters")
```

```python
sc.tl.draw_graph(adata, layout='fr')
print("Force-directed graph (FR) done")
sc.pl.draw_graph(adata, layout='fr', color = "seurat_clusters")
```

```python
sc.tl.draw_graph(adata, layout='kk')
print("Force-directed graph (KK) done")
sc.pl.draw_graph(adata, layout= "kk", color = "seurat_clusters")
```

```python
sc.tl.diffmap(adata)
print("Diffusion maps done")
sc.pl.diffmap(adata, color = "seurat_clusters")
```
