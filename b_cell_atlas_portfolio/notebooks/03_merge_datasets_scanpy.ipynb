```python
import os
from pathlib import Path

# Override with: export B_CELL_ATLAS_DATA=/path/to/your/data
DATA_ROOT = Path(os.environ.get('B_CELL_ATLAS_DATA', 'data'))

import csv
import gc
import scanpy as sc
import anndata as ad
```

```python
import anndata as ad
import os

files = [
str(DATA_ROOT / "geo" / "GSE113660_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE132509_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE147766_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE148218_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE154109_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE166345_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE174376_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE192906_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE216155_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE220651_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE220946_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE227122_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE227983_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE230259_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE231859_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE235063_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE235923_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE236351_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE253865_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE261957_final_csr_corrected.h5ad"),
str(DATA_ROOT / "geo" / "GSE262271_final_csr_corrected.h5ad")
]

# making sure all files exist
for file in files:
    if not os.path.exists(file):
        print(f"Warning: File not found: {file}")

# Extract dataset identifiers from filenames for better labeling
keys = [os.path.basename(file).split('_')[0] for file in files]
print(f"Found {len(files)} files with keys: {keys}")

# processing one file at a time
adatas = []
for file, key in zip(files, keys):
    try:
        adata = ad.read_h5ad(file)
        adata.obs_names_make_unique()
        adata.var_names = adata.var_names.astype(str)
        adata.var_names_make_unique()
        adatas.append(adata)
        print(f"Successfully loaded {key} with {adata.n_obs} cells and {adata.n_vars} genes")
    except Exception as e:
        print(f"Error loading {file}: {e}")

# concat
merged_adata = ad.concat(
    adatas,
    axis=0,
    join="outer",  
    label="dataset",
    keys=keys,
)
merged_adata.obs_names_make_unique()

print(f"Merged dataset has {merged_adata.n_obs} cells and {merged_adata.n_vars} genes")
```
