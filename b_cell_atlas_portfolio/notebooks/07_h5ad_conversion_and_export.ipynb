```python
## if the files are counts and features only
# File paths
counts_file = str(DATA_ROOT / "raw" / "gse218003" / "GSM6745506_M017AAB_T1_scR_CEL-Seq2_counts.tsv.gz")
features_file = str(DATA_ROOT / "raw" / "gse218003" / "GSM6745506_M017AAB_T1_scR_CEL-Seq2_fixed.features.tsv.gz")

# Load counts matrix
counts = pd.read_csv(counts_file, sep="\t", index_col=0)

# Load features (gene metadata), skipping the first row
features = pd.read_csv(features_file, sep="\t", header=None, skiprows=1)
features.columns = ["gene_id", "gene_name"]

# Ensure consistency between counts and features
counts.index = features["gene_name"]  # Use gene names as row indices

# Create AnnData object
adata = sc.AnnData(X=counts.T)  # Transpose so cells are rows and genes are columns

# Add gene metadata to .var
adata.var["gene_id"] = features["gene_id"].values

# Optional debugging output
# print(f"Counts rows: {counts.shape[0]}, Features rows: {features.shape[0]}")
# print("Features tail:")
# print(features.tail())
# if not features["gene_name"].is_unique:
#     print("Duplicate gene names found!")
#     features = features.drop_duplicates(subset="gene_name")
```
